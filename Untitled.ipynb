```python
import pandas as pd
```

```python
df = pd.read_csv("_annotations.csv")
```

```python
df.rename(columns={'class': 'classes'}, inplace=True)
```

```python
a=[]
b=[]
d=[]
e=[]
#c=0
for x in range(31356):
    if df['classes'][x] == 'bottle' :
        df['filename'][x] 
        a.append(x)
        
    elif df['classes'][x] == 'person' :
        df['filename'][x] 
        b.append(x)
        
    elif df['classes'][x] == 'chair' :
        df['filename'][x] 
        d.append(x)
    elif df['classes'][x] == 'cell phone' :
        df['filename'][x] 
        e.append(x)
        #c=c+1

#print(c)
#868 bottle
#9546 person
#1450 chair
#246 cellphone
```

```python
merged_list = list(set(a + b  + d + e))
print(merged_list)
```

```python
len(merged_list)
```

```python
file=[]
wid=[]
hei=[]
clas=[]
xmi=[]
ymi=[]
xma=[]
yma=[]
for x in range(12110):
    file.append(df.filename[merged_list[x]])
    wid.append(df.width[merged_list[x]])
    hei.append(df.height[merged_list[x]])
    clas.append(df.classes[merged_list[x]])
    xmi.append(df.xmin[merged_list[x]])
    ymi.append(df.ymin[merged_list[x]])
    xma.append(df.xmax[merged_list[x]])
    yma.append(df.ymax[merged_list[x]])
```

```python
df_new = pd.DataFrame({'filename': file,
                       'width': wid,
                       'height': hei,
                       'class': clas,
                       'xmin': xmi,
                       'ymin': ymi,
                       'xmax' : xma,
                       'ymax' : yma})
```
